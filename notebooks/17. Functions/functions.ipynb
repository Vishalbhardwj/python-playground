```python
def greet(name = '-----'):
    print('Hello',name)
```

```python
greet('Ashish')
greet(name = 'Ashish')
greet()
print('-'*20)
```

```python
# Calling Different Functions
lst = [1,2,3,4,5]
```

```python
def sq(lst):
    return [i**2 for i in lst]
def cu(lst):
    return [i**3 for i in lst]
```

```python
def final(lst):

    lst_1 = sq(lst)
    lst_2 = cu(lst)

    return [lst_1[i] + lst_2[i] for i in range(len(lst_1))]
```

```python
print(final(lst))
```

1. Arithmetic Operators

```python
# Addition
print(2 + 3)
print(2.5 + 3.5)
print(2.5 + 3)
print(True + False)
print('Ashish ' + 'Jangra' + ' is a data science mentor')
print('--------')
```

```python
# Subtraction
print(2 - 3)
print(2.5 - 3.5)
print(2.5 - 3)
print(False - True)
print('--------')
```

```python
# Multiplication
print(2 * 3)
print(2.5 * 3.5)
print(2.5 * 3)
print(True * True)
print('--------')
```

```python
# Division
print(3 / 2)
print(2.5 / 5.0)
print(2.5 / 1)
print('--------')
```

```python
# Floor Division
print(3 // 2)
print(2.5 // 5.0)
print(2.5 // 1)
print('--------')
```

```python
# Modulo
print(3 % 2)
print(2.5 % 5.0)
print(2.5 % 1)
print('--------')
```

```python
# Exponentiation
print(3 ** 5)
print(2.5 ** 5.0)
print(2.5 ** 1)
print('--------')
```

```python
# abs
print(abs(-3.5))
print(abs(-35))
print('--------')
```

```python
# round
print(round(3.42349238, 4))
print('--------')
```

```python
# pow
print(pow(3,5))
print(pow(5,3))
print('--------')
```

2. Comparison Operators

```python
# ==
print(1 == 2)
print(1.0 == 1)
print('A' == 'a')
print(True == 1)
print(False == 0)
print('--------')
```

```python
# !=
print(1 != 2)
print(1.0 != 1)
print('A' != 'a')
print(True != 1)
print(False != 0)
print('--------')
```

```python
# >
print(1 > 2)
print(1.0 > 1)
print(3 > True)
print(False > 0)
print('--------')
```

```python
# <
print(1 < 2)
print(1.0 < 1)
print(3 < True)
print(False < 0)
print('--------')
```

```python
# <=
print(1 <= 2)
print(1.0 <= 1)
print(3 <= True)
print(False <= 0)
print('--------')
```

```python
# >=
print(1 >= 2)
print(1.0 >= 1)
print(3 >= True)
print(False >= 0)
print('--------')
```

3. Logical Operators

```python
# and
print(True and True)
print(False and True)
print(True and False)
print(False and False)
print('--------')
```

```python
# or
print(True or True)
print(False or True)
print(True or False)
print(False or False)
print('--------')
```

```python
# not
print(not True)
```

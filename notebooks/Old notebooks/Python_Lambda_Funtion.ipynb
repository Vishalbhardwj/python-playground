## Python Lambda funtion!

```python
# Normal funtion

def addition(a,b):
    return a+b
```

```python
addition(4,5)
```

```python
addition = lambda a,b: a+b
```

```python
addition(12,45)
```

```python
def even_or_odd(num):
    if(num%2==0):
        return True
    else:
        return False
```

```python
even_or_odd(45)
```

```python
even_or_odd(22)
```

```python
Even_or_odd = lambda num: num%2==0 
```

```python
Even_or_odd(22)
```

```python
Even_or_odd(77)
```

```python
def addition(x,y,z):
    return x+y+z
```

```python
addition(34,56,78)
```

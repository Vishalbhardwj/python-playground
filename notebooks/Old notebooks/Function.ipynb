```python


def name_age(name , age = 22):
    return "My name is {} and age is {}".format(name,age)
```

```python
print(name_age)
```

```python
val = name_age("vishal")
```

```python
print(val)
```

```python
def names_ages(*args,**kwargs):
    print(args)
    print(kwargs)
```

```python
names_ages('vishal','awadhesh',age =22,ages =23)
```

```python
lst = ['rohit','yash']
dic = {'age' : 21,'ages' :23}
names_ages(*lst,**dic)
```

```python
lst1= [1,2,3,6,8,4,5,7,8,10,43,42,56,78,2,22,67]

def sumOfEvenOddNo(lst1):
    sumOfEven = 0
    sumOfOdd = 0
    for i in lst1:
        if(i%2==0):
            sumOfEven+=i
        else:
            sumOfOdd+=i
    return sumOfEven,sumOfOdd


print(sumOfEvenOddNo(lst1))          
    
```

```python
val2 = sumOfEvenOddNo(lst1)
print(val2)
```

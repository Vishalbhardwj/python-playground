# Python List Comprehension & String Formatting 

```python
lst=[]
def list_square(list1):
    for i in list1:
        lst.append(i*i)
    return lst
        
```

```python
lst2 = [1,2,3,4,5,6,7,8]
```

```python
print(list_square(lst2))
```

```python
[ i*i  for i in lst2 ]
```

```python
[ i*i for i in lst2 if(i%2==0)]    # List Comprehension!
```

```python
[ i*i for i in lst2 if(i%2!=0)]
```

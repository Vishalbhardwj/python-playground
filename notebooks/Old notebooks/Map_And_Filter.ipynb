# Map & Filter in python!

```python
def even_or_odd(num):
    if(num%2==0):
        return True
    else:
        return False
```

```python
list1 = [23,45,46,22,12,18,17,11,13,19,50,43]
```

```python
for i in list1:
     print(even_or_odd(i))
```

```python
list(map(even_or_odd,list1))    #Map
```

```python
list(filter(even_or_odd,list1))   #Filter 
```

```python
list(filter(lambda num : num%2!=0,list1))    #All odd number present in the list!
```
